```python
#Q2 Sensor using numpy
import numpy as np
#load csv via list of lists
file_path = "sensor_data.csv"

data_list = []
with open(file_path, 'r') as f:
    for line in f:
        row = line.strip().split(',')
        data_list.append(row)
print("loaded rows:", len(data_list))
print("sample row:", data_list[0][:10])
```

```python
#given list to numpy array
data_array = np.array(data_list, dtype=float)
print("shape of array:", data_array.shape)
```

```python
#required data cleaning
data_array[data_array == -999] = np.nan
data_array[data_array < 0] = np.nan
data_array[data_array > 100] = np.nan

print("after cleaning, sample row:")
print(data_array[0][:10])
```

```python
#required data extraction
#columnwise
col_mean = np.nanmean(data_array, axis=0)
#rowwise
row_median = np.nanmedian(data_array, axis=1)
#most invalid sensor
nan_count = np.sum(np.isnan(data_array), axis=0)
worst_sensor = np.argmax(nan_count)
print("avg moisture per sensor:", col_mean[:10])
print("median moisture per hour:", row_median[:10])
print("most invalid read sensor:", worst_sensor)
```

```python
#min-maxing
valid_min = np.nanmin(data_array)
valid_max = np.nanmax(data_array)
Xnor = (data_array - valid_min) / (valid_max - valid_min)
print("normal row:")
print(Xnor[0][:10])
```

```python
#new csv for our data
output_file = "sensor_data_normalized.csv"
np.savetxt(output_file, Xnor, fmt='%.18e', delimiter=' ', newline='\n', header='Normalized Sensor Data', footer='', comments='# ', encoding=None)
print("normalized data saved to:", output_file)
```
