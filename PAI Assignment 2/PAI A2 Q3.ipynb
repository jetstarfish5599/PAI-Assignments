```python
#Q3 titanic dataset using pandas
import pandas as pd
#load dataset
file = "Titanic-Dataset.csv"
df = pd.read_csv(file)
#check columns and missing 
print("dataset info:\n")
for col in df.columns:
    dtype = str(df[col].dtype)
    missing = df[col].isnull().mean() * 100
    print(f"{col:15} type={dtype:<10} missing={missing:.2f}%")

```

```python
#missing data manipulation
df['Age'] = df.groupby(['Pclass', 'Sex'])['Age'].transform(lambda x: x.fillna(x.median()))
mode_val = df['Embarked'].mode()[0]
df['Embarked'].fillna(mode_val, inplace=True)
df.drop(columns=['Cabin'], inplace=True)
```

```python
#feature family or alone
df['FamilySize'] = df['SibSp'] + df['Parch']
df['IsAlone'] = df['FamilySize'].apply(lambda x: 1 if x == 0 else 0)
```

```python
#Age to int64
df['Age'] = df['Age'].astype('int64')
```

```python
#new csv for our data
output_file = "titanic_cleaned.csv"
df.to_csv(output_file, index=False)
print("cleaned data saved to:", output_file)
```
