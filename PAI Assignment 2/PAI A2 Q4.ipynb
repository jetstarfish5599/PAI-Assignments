```python
#Q4 hypotesis using pandas 
import pandas as pd
#load datasets from Q3
titanic = pd.read_csv("titanic_cleaned.csv")
fares = pd.read_csv("ticket_fares.csv")
print("titanic shape:", titanic.shape)
print("fares shape:", fares.shape)
```

```python
#merge datasets
merged = pd.merge(titanic, fares, on='Ticket', how='left')
print("merged shape:", merged.shape)
print("columns after merge:\n", merged.columns.tolist()[:10])
```

```python
#Hypothesis 1
bins = [0, 12, 19, 59, 120]
labels = ['Child', 'Teen', 'Adult', 'Senior']
merged['AgeGroup'] = pd.cut(merged['Age'], bins=bins, labels=labels, right=False)


h1 = merged.groupby(['Sex', 'AgeGroup'])['Survived'].mean().reset_index()

print("hypothesis 1 result:")
print(h1)
```

```python
import pandas as pd

#Hypothesis 2
titanic = pd.read_csv("titanic_cleaned.csv")
df = titanic
h2a = df.groupby('Pclass')['Survived'].mean().reset_index()
df['FareBin'] = pd.qcut(df['Fare'], 4, labels=['Low', 'Medium', 'High', 'Very High'])
h2b = df.groupby('FareBin')['Survived'].mean().reset_index()
#Print
print("\nHypothesis 2:Class based")
print(h2a)
print("\nHypothesis 2:Fare quantity")
print(h2b)
```

```python
#result in report.txt
with open("report.txt", "w") as f:
    f.write("Hypothesis 1:women and children 1st\n")
    f.write(str(h1))
    f.write("\n\nHypothesis 2:wealthy survived more\n")
    f.write("method A:\n")
    f.write(str(h2a))
    f.write("\n\n method B:\n")
    f.write(str(h2b))
    f.write("\n\nConclusion: Wealthier passengers  show higher survival, supporting the hypothesis.")

print("report.txt generated successfully.")
```
